```python
from rrtl.models.decoders import (
    Decoder,
)
from rrtl.models.encoders import (
    TokenLevelEncoder,
    SentenceLevelEncoder,
)
import sys
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.distributions.relaxed_bernoulli import RelaxedBernoulli
from torch.distributions.bernoulli import Bernoulli
from torch.distributions.kl import kl_divergence

from transformers import BertModel

class FullContextSentimentModel(nn.Module):
    def __init__(self, args):
        super(FullContextSentimentModel, self).__init__()
        self.args = args
        self.decoder = Decoder(args)
    
    def forward(self, batch, mode='train'):
        input_ids = batch['input_ids']
        attention_mask = batch['attention_mask']
        labels = batch['labels']
        output = self.decoder(input_ids, attention_mask, labels)

        loss = output.loss
        return {
            'loss': loss,
            'logits': output.logits,
        }

    def forward_eval(self, batch):
        with torch.no_grad():
            return self.forward(batch, mode='eval')
```

```python
def eval_prob(output):
        logits=output['logits'][0]
        #print(logits)
        exp_logits=torch.exp(logits)
        prob=exp_logits[1]/torch.sum(exp_logits)
        return prob

def eval_pns1(batch,output):
    for t in batch['input_ids']:
        print(tokenizer.convert_ids_to_tokens(t))
    attention_mask=batch['attention_mask'][0]
    length=torch.sum(attention_mask)
    batch_new=batch
    prob1=eval_prob(output)
    #print('prob1:',prob1)
    logpns=0
    for i in range(length):
        attention_mask_new=attention_mask.clone()
        attention_mask_new[i]=0          
        batch_new['attention_mask']=torch.unsqueeze(attention_mask_new,0)
        output_new = model.forward_eval(batch_new)
        prob2=eval_prob(output_new)
        print('prob2:',prob2)
        logpns+=torch.log(abs(prob1-prob2))/length
    return logpns

def eval_pns2(batch,output):
    attention_mask=batch['attention_mask'][0]
    length=torch.sum(attention_mask)
    batch_new=batch
    prob1=eval_prob(output)
    print('prob1:',prob1)
    logpns=0
    for i in range(length):
        attention_mask_new=attention_mask.clone()
        attention_mask_new[i]=0          
        batch_new['attention_mask']=torch.unsqueeze(attention_mask_new,0)
        output_new = model.forward_eval(batch_new)
        prob2=eval_prob(output_new)
        print('prob2:',prob2)
        logpns+=torch.log(abs(prob1-prob2))/length
    return logpns
```
