```python
import os
import sys
from utils.spark_init import spark_session
from pyspark.sql.functions import *
import pandas as pd
import numpy as np
from itertools import chain
from __future__ import print_function
from utils.extract import diag_ext,diag_clm_ext,prod_clm_ext,clm_qc
from datetime import datetime
from time import time
from pyspark.sql import Window

spark = spark_session()

HDFS_PATH = '/user/twu2/Iveric_GA_intern_project_202207'
pos_dx_path = '{}/clm_data/pos_dx_clm.parquet'.format(HDFS_PATH)
risk_dx_path = '{}/clm_data/risk_dx_clm.parquet'.format(HDFS_PATH)
pos_dt_path = '{}/clm_data/pos_index_dt.parquet'.format(HDFS_PATH)
risk_dt_path = '{}/clm_data/risk_index_dt.parquet'.format(HDFS_PATH)

pos_dx_clm = spark.read.parquet(pos_dx_path)
risk_dx_clm = spark.read.parquet(risk_dx_path)
pos_dt = spark.read.parquet(pos_dt_path)
risk_dt = spark.read.parquet(risk_dt_path)

v_diag = spark.sql("""
  select distinct diag_cd
  from prod_df2_us9.v_diagnosis
""")

cols = ['patient_id','svc_fr_dt','diag_cd',
        'rendering_provider_id','provider_pri_spcl_cd']

######################
# positive cohort claim data
######################

pos_dx_clm_flt = pos_dx_clm.select(cols).dropDuplicates()

pos_dx_clm_flt = pos_dx_clm_flt\
  .join(pos_dt, on='patient_id', how='inner')\
  .filter(col('svc_fr_dt').cast('date') < col('index_dt'))\
  .filter(col('svc_fr_dt').cast('date') >= col('lkbk_dt'))\
  .dropDuplicates()
  
pos_dx_clm_df = pos_dx_clm_flt\
  .join(v_diag, 'diag_cd', 'inner')\
  .withColumn('svc_dt', col('svc_fr_dt').cast('date'))\
  .select('patient_id','svc_dt','diag_cd',
          'rendering_provider_id','provider_pri_spcl_cd')\
  .toPandas()
  
pos_dx_clm_df = pos_dx_clm_df.sort_values(by=['patient_id','svc_dt'])

pos_dx_clm_df.to_parquet('/home/cdsw/internship_project/data/pos_dx.parquet', index=False)

pos_dx_clm_df = pd.read_parquet('/home/cdsw/internship_project/data/pos_dx.parquet')

######################
# risk cohort claim data
######################

risk_dx_clm_flt = risk_dx_clm\
  .select(cols).dropDuplicates()\
  .join(risk_dt, on='patient_id', how='inner')\
  .join(v_diag, 'diag_cd', 'inner')\
  .withColumn('svc_dt', col('svc_fr_dt').cast('date'))\
  .filter(col('svc_dt') < col('index_dt'))\
  .filter(col('svc_dt') >= col('lkbk_dt'))\
  .select('patient_id','svc_dt','diag_cd',
          'rendering_provider_id','provider_pri_spcl_cd')

#risk_dx_clm_flt.write.mode('overwrite')\
#  .parquet(HDFS_PATH + '/clm_data/risk_dx_clm_filtered.parquet')
  
#risk_dx_clm_flt = spark.read.parquet(HDFS_PATH + '/clm_data/risk_dx_clm_filtered.parquet')
risk_dx_clm_flt.write.mode('overwrite')\
  .parquet(HDFS_PATH + '/clm_data/risk_dx_clm_filtered.parquet')
  
risk_dx_clm_flt = spark.read.parquet(HDFS_PATH + '/clm_data/risk_dx_clm_filtered.parquet')
```
