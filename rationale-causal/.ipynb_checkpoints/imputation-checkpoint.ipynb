# Load Model

```python
from transformers import AutoConfig,AutoModel,PreTrainedModel
class CustomModel(PreTrainedModel):
    def __init__(self, config):
        super().__init__(config)
        self.backbone = AutoModel.from_config(config)

        self.output = nn.Linear(config.hidden_size, 1)
        self.output1 = nn.Linear(120, 1)
        self.sig= nn.Sigmoid()

    def forward(
        self,
        input_ids,
        attention_mask=None,
        token_type_ids=None,
        position_ids=None,
        labels=None,
    ):
        outputs = self.backbone(
            input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            position_ids=position_ids,
        )
        
        sequence_output = outputs.last_hidden_state
        outputs = self.output(sequence_output)
        outputs=torch.permute(outputs, (0, 2, 1))
        outputs = self.output1(outputs)
        outputs=torch.squeeze(outputs)
        outputs=self.sig(outputs)

            
        return {
            "outputs": outputs
        }
def pad(args,seq):
        return seq + (args.max_length - len(seq)) * [0]
```

# Load Data

```python

def imputation(aspect,args):

    checkpoint = torch.load('/home/ec2-user/SageMaker/Causal-Rationale/rationale-causal/experiments/beer/imputation/Look_score_model.pt', map_location='cpu')
    states = checkpoint['states']
    model_name = "bert-base-uncased"
    cfg = AutoConfig.from_pretrained(model_name)
    model_look= CustomModel(cfg)
    model_look=model_look
    model_look.load_state_dict(states)
    model_look.eval()


    checkpoint = torch.load('/home/ec2-user/SageMaker/Causal-Rationale/rationale-causal/experiments/beer/imputation/Aroma_score_model.pt', map_location='cpu')
    states = checkpoint['states']
    model_name = "bert-base-uncased"
    cfg = AutoConfig.from_pretrained(model_name)
    model_aroma= CustomModel(cfg)
    model_aroma=model_aroma
    model_aroma.load_state_dict(states)
    model_aroma.eval()

    checkpoint = torch.load('/home/ec2-user/SageMaker/Causal-Rationale/rationale-causal/experiments/beer/imputation/Palate_score_model.pt', map_location='cpu')
    states = checkpoint['states']
    model_name = "bert-base-uncased"
    cfg = AutoConfig.from_pretrained(model_name)
    model_palate= CustomModel(cfg)
    model_palate=model_palate
    model_palate.load_state_dict(states)
    model_palate.eval()

    mode='train'
    path = config.DATA_DIR / f'sentiment/data/source/beer_{aspect}.{mode}_120'
    df= pd.read_csv(path, delimiter='\t')

    if aspect=='Look':
        df.rename(columns = {'label':'label_0'}, inplace = True)
        df['label_1']=[1 for i in range(len(df))]
        df['label_2']=[1 for i in range(len(df))]                                
    if aspect=='Aroma':
        df.rename(columns = {'label':'label_1'}, inplace = True)
        df['label_0']=[1 for i in range(len(df))]
        df['label_2']=[1 for i in range(len(df))]   
    if aspect=='Palate':
        df.rename(columns = {'label':'label_2'}, inplace = True)
        df['label_0']=[1 for i in range(len(df))]
        df['label_1']=[1 for i in range(len(df))]   


    tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased', cache_dir=args.cache_dir)
    for ind in tqdm(df.index):
            text =df['text'][ind]
            tokenized = tokenizer.encode_plus(text, add_special_tokens=True)
            input_ids = tokenized['input_ids']
            attention_mask = tokenized['attention_mask']
            if  len(input_ids) >= args.max_length:
                    input_ids = input_ids[:args.max_length - 1] + [102]
                    attention_mask = attention_mask[:args.max_length - 1] + [1]
            else:
                    input_ids = input_ids
                    attention_mask = attention_mask 
            input_ids = pad(args,input_ids)
            attention_mask = pad(args,attention_mask)

            if aspect=='Look':
                output_aroma=model_aroma(input_ids=torch.unsqueeze(torch.tensor(input_ids),0),attention_mask=torch.unsqueeze(torch.tensor(attention_mask),0))
                df['label_1'][ind]=output_aroma['outputs'].item()
                output_palate=model_palate(input_ids=torch.unsqueeze(torch.tensor(input_ids),0),attention_mask=torch.unsqueeze(torch.tensor(attention_mask),0))
                df['label_2'][ind]=output_palate['outputs'].item()

            elif aspect=='Aroma':
                output_look=model_look(input_ids=torch.unsqueeze(torch.tensor(input_ids),0),attention_mask=torch.unsqueeze(torch.tensor(attention_mask),0))
                df['label_0'][ind]=output_look['outputs'].item()
                output_palate=model_palate(input_ids=torch.unsqueeze(torch.tensor(input_ids),0),attention_mask=torch.unsqueeze(torch.tensor(attention_mask),0))
                df['label_2'][ind]=output_palate['outputs'].item()

            elif aspect=='Palate':
                output_look=model_look(input_ids=torch.unsqueeze(torch.tensor(input_ids),0),attention_mask=torch.unsqueeze(torch.tensor(attention_mask),0))
                df['label_0'][ind]=output_look['outputs'].item()
                output_aroma=model_aroma(input_ids=torch.unsqueeze(torch.tensor(input_ids),0),attention_mask=torch.unsqueeze(torch.tensor(attention_mask),0))
                df['label_1'][ind]=output_aroma['outputs'].item()
    
    print('finished')

    return df





```

```python
df_look=imputation('Look',args)
df_aroma=imputation('Aroma',args)
df_palate=imputation('Palate',args)
```
